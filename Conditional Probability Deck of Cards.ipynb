# Problem: Calculate the probability of drawing a red card from the deck of cards given that the card drawn is a face card (jack, queen, or king). 

## P(Red | Face) which is equivalent to P(Red ∩ Face) / P(Face)

## <span style = "color:red"> P(Red | Face) = P(Red ∩ Face) / P(Face) </span>
##              = (red_face_cards / total_cards) / (face_cards / total_cards)
##              = red_face_cards / face_cards


```python
# Total number of cards in the deck
total_cards = 52

# Number of red cards in the deck
red_cards = 26

# Number of face cards in the deck
face_cards = 12

# Number of face cards that are red
red_face_cards = 6  # There are 2 red jacks, 2 red queens, and 2 red kings

# Calculate conditional probability: P(Red | Face)
conditional_prob = red_face_cards / face_cards

print(f"Conditional Probability of drawing a red card given a face card: {conditional_prob:.4f}")
```

# If you draw two cards out of a deck of cards one after the another what is the probability that both are spades ?

## Sure, let's approach this using conditional probability, where:

## <span style = "color:red"> Event A: First card is a spade. </span>
## <span style = "color:red"> Event B: Second card is a spade. </span>
    
## <span style = "color:blue"> Calculating P(A): </span>
## The probability of drawing a spade on the first draw is 13/52, as there are 13 spades in a deck of 52 cards.

## <span style = "color:blue"> Calculating P(B|A): </span>
## After drawing a spade on the first draw, there are now 51 cards left in the deck, and 12 of them are spades.

## So, the probability of drawing a spade on the second draw, given that the first card drawn was a spade, is 12/51.

## P(B|A) = P(A ∩ B) / P(A) 

## <span style = "color:red"> P(A ∩ B) is the probability of both events A and B occurring. </span>


## P(A ∩ B) = P(B|A) * P(A)
##         = 12/51 * 13/52
##         = 0.0588

## <span style = "color:green"> The probability that both the cards drawn one after another are spades is around 5.9 % </span>


# Problem: What is the total probability of drawing two cards , an ace of clubs and a red card ? 

## To calculate the total probability of drawing both an Ace of Clubs and a red card in two consecutive draws from a standard deck of 52 playing cards, we need to consider all possible cases. This can be done using the law of total probability, which accounts for various ways an event can occur based on different conditions.

## Let's denote the events:

## <span style = "color:red"> Event A: Drawing an Ace of Clubs.</span>
## <span style = "color:red"> Event B: Drawing a red card (hearts or diamonds).</span>

## We want to calculate the probability of drawing both an Ace of Clubs and a red card in two consecutive draws. This can occur in two ways:

## Drawing the Ace of Clubs first, then a red card.
## Drawing a red card first, then the Ace of Clubs.
## We'll calculate the probabilities of these two cases separately and then add them to find the total probability.

## <span style = "color:red"> Case 1: Ace of Clubs first, then a red card:</span>

## <span style = "color:blue"> P(A ∩ B) = P(A) * P(B|A) </span>

## Probability of drawing the Ace of Clubs: 1/52 (since there's only one Ace of Clubs in the deck).
## After drawing the Ace of Clubs, there are 51 cards remaining, out of which 26 are red cards.
## Probability of drawing a red card after the Ace of Clubs: 26/51.
## So, the probability of drawing an Ace of Clubs first and then a red card is: (1/52) * (26/51).

## <span style = "color:red"> Case 2: Red card first, then the Ace of Clubs: </span>

## <span style = "color:blue"> P(A ∩ B) = P(B) * P(A|B) </span>

## Probability of drawing a red card: 26/52 (since there are 26 red cards in the deck).
## After drawing a red card, there are 51 cards remaining, out of which 1 is the Ace of Clubs.
## Probability of drawing the Ace of Clubs after a red card: 1/51.
## So, the probability of drawing a red card first and then the Ace of Clubs is: (26/52) * (1/51).

## Total Probability:
## <span style = "color:red"> Now, we add the probabilities of the two cases to get the total probability: </span>

## Total Probability = Probability of Case 1 + Probability of Case 2
## = (1/52) * (26/51) + (26/52) * (1/51)
## = 26/2652 + 26/2652
## = 52/2652
## ≈ 0.0196

## <span style = "color:green"> The total probability of drawing both an Ace of Clubs and a Red card in two consecutive draws is approximately 0.0196, or around 2% </span>
