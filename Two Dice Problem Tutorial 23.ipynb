# Select random names from a given list of names

### random.choice() and random.sample()

```python
import random

# Import the random Module: We start by importing the random module, 
# which provides functions for generating random numbers and performing random operations.

names = ["Alice", "Bob", "Charlie", "David", "Eve"]

# List of Names: We define a list called names that contains several names as strings.

random_name = random.choice(names)
print(f"Randomly selected name: {random_name}")

# Randomly Select a Name: We use the random.choice() function to select a single 
# random name from the names list. This function takes a sequence (like a list) 
# as an argument and returns a randomly selected element from that sequence.

sample_size = 2
random_sample = random.sample(names, sample_size)
print("Random sample of names:", random_sample)

# Random Sample of Names: We use the random.sample() function to generate a random sample 
# of names from the names list. This function takes two arguments: the sequence from which 
# to sample (in this case, the names list), and the sample size (in this case, sample_size is 
# set to 2). It returns a list of unique randomly selected elements from the given sequence.
```

# Random Variables and Discrete Probability

# A python code to represent the discrete probability as a histogram

```python
import matplotlib.pyplot as plt

# The calculate_discrete_probabilities() function calculates the probabilities 
# of all possible outcomes when two dice are rolled together.
def calculate_discrete_probabilities():
    
    outcomes = [(i, j) for i in range(1, 7) for j in range(1, 7)]
# outcomes = [(i, j) for i in range(1, 7) for j in range(1, 7)]:
# This line uses a nested list comprehension to generate all possible pairs of outcomes
# when rolling two dice. It iterates over values of i and j from 1 to 6, 
# creating tuples (i, j) for each combination. This represents the sample space 
# of all possible dice roll outcomes.
    
    probabilities = {}
# probabilities = {}:
# Here, an empty dictionary named probabilities is created. 
# This dictionary will store the counts of each possible sum of two dice rolls.
# Like 2 occurs 1 time (1,1), 3 occors 2 times (2,1) and (1,2), 4 occors 3 times (1,3),(3,1),(2,2)

    total_outcomes = len(outcomes)
# total_outcomes = len(outcomes):
# The variable total_outcomes is assigned the number of possible outcomes. 
# In this case, since we're considering two dice with 6 sides each, 
# there are 6 * 6 = 36 possible outcomes.
    
    for outcome in outcomes:
# for outcome in outcomes::
# This loop iterates through each possible outcome (pair of dice rolls) 
# generated in the outcomes list.

        dice_sum = sum(outcome)
# dice_sum = sum(outcome):
# For each outcome, the sum() function is used to calculate the sum of the two dice rolls.
# This dice_sum represents the value of interest for our random variable.

        if dice_sum in probabilities:
# if dice_sum in probabilities::
# This if statement checks if the calculated dice_sum is already a key 
# in the probabilities dictionary.

            probabilities[dice_sum] += 1
# probabilities[dice_sum] += 1:
# If the dice_sum is already in the dictionary, this line increments its value by 1. 
# This keeps track of how many times each sum has been encountered.

        else:
            probabilities[dice_sum] = 1
# If the dice_sum is not in the dictionary, 
# this line adds it as a new key with a value of 1. This initializes the count for that sum.
            
    for dice_sum, count in probabilities.items():
# for dice_sum, count in probabilities.items()::
# This loop iterates through the items (key-value pairs) in the probabilities dictionary.

        probabilities[dice_sum] = count / total_outcomes
# probabilities[dice_sum] = count / total_outcomes:
#For each sum in the dictionary, this line calculates the probability 
# by dividing the count of occurrences by the total number of outcomes. 
# This normalization step ensures that the probabilities sum up to 1.
        
    return probabilities
# Finally, the function returns the probabilities dictionary containing the discrete
# probability distribution for the sum of two dice rolls.

# The main() function calls calculate_discrete_probabilities() to get the probabilities.
def main():
    probabilities = calculate_discrete_probabilities()
# We use plt.bar() from the matplotlib.pyplot module to create a bar plot (histogram)
# representing the discrete probability distribution.
    plt.bar(probabilities.keys(), probabilities.values(), align='center')
# The align='center' parameter ensures that the bars are centered over 
# their corresponding x-values.
    plt.xlabel('Sum of Two Dice')
    plt.ylabel('Probability')
    plt.title('Discrete Probability Distribution of Two Dice Rolls')
    plt.xticks(range(2, 13))
# We set the x-axis labels using plt.xticks() to match the possible
# sums of two dice rolls (from 2 to 12).
    plt.show()
# We display the histogram using plt.show()

if __name__ == "__main__":
    main()
```

# When two dice are rolled together calculate the Random Variables (data points) of the sum, the variance and the standard deviation

```python
def calculate_discrete_probabilities():
    outcomes = [(i, j) for i in range(1, 7) for j in range(1, 7)]
    
    probabilities = {}
    total_outcomes = len(outcomes)
    
    for outcome in outcomes:
        dice_sum = sum(outcome)
        if dice_sum in probabilities:
            probabilities[dice_sum] += 1
        else:
            probabilities[dice_sum] = 1
            
    for dice_sum, count in probabilities.items():
        probabilities[dice_sum] = count / total_outcomes
        
    return probabilities

def calculate_mean(probabilities):
    mean = sum(dice_sum * prob for dice_sum, prob in probabilities.items())
    return mean
# This function calculates the mean (also known as the expected value) of a 
# random variable based on its probability distribution. In the context of rolling two dice,
# the mean represents the average value of all possible sums that can occur.
# probabilities: This is the dictionary containing the discrete probability 
# distribution of the sums of two dice rolls.
# dice_sum and prob: These variables are unpacked from the items 
# in the probabilities dictionary. dice_sum represents a possible sum of two dice rolls,
# and prob represents the probability associated with that sum.
# The function calculates the mean using the formula:
# mean =  [ (dice_sum) * (prob) ]
# It iterates through each possible sum in the probabilities dictionary, 
# multiplies the sum by its corresponding probability, and adds up these products 
# to get the mean.

def calculate_variance(probabilities, mean):
    variance = sum(prob * (dice_sum - mean) ** 2 for dice_sum, prob in probabilities.items())
    return variance

# This function calculates the variance of a random variable. 
# Variance measures how much the values of the random variable differ from their mean.
# In the context of rolling two dice, the variance measures how spread out 
# the possible sums are around the mean.
# probabilities: This is the same dictionary containing the discrete probability distribution.
# mean: This is the mean calculated using the previous function.
# The function calculates the variance using the formula:
# variance =  [ (prob) * (dice_sum - mean)^2 ]
# It iterates through each possible sum in the probabilities dictionary, 
# calculates the squared difference between the sum and the mean, 
# multiplies it by the corresponding probability, 
#and adds up these products to get the variance.

def calculate_standard_deviation(variance):
    return variance ** 0.5

#This function calculates the standard deviation, which is the square root of the variance. 
# The standard deviation gives us an idea of how spread out the values of the random variable
# are from the mean.
# variance: This is the variance calculated in the previous function.
# The standard deviation is calculated by taking the square root of the variance.

def main():
    probabilities = calculate_discrete_probabilities()
    mean = calculate_mean(probabilities)
    variance = calculate_variance(probabilities, mean)
    standard_deviation = calculate_standard_deviation(variance)
    
    print("Discrete Probability Distribution:")
    for dice_sum, prob in probabilities.items():
        print(f"Sum: {dice_sum}, Probability: {prob:.4f}")
    
    print("\nMean:", mean)
    print("Variance:", variance)
    print("Standard Deviation:", standard_deviation)

if __name__ == "__main__":
    main()

# The main() function calculates the discrete probability distribution, 
# mean, variance, and standard deviation, and then prints out the results.
```
