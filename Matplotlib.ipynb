# Multi-series line plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of values for x coordinates
x = np.linspace(0, 2*np.pi, 100)  # Generate x values from 0 to 2*pi
x1 = x / (2*np.pi)


# In this section, an array x is created using NumPy's linspace function. 
# The linspace function generates an evenly spaced array of numbers. 
# In this case, it generates 100 values between 0 and 2 (two times , approximately 6.28318). 
# The second line creates another array x1 where each value of x is divided by 2. 
# This is done to map the x-values to the range [0, 1], which will later be used for the x-axis of the plot.

# Compute the corresponding cosine, sine, and tangent values
cos_values = np.cos(x)
sin_values = np.sin(x)
tan_values = np.tan(x)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the cosine, sine, and tangent curves in different colors
ax.plot(x1, cos_values, label='Cosine Curve', color='red')
ax.plot(x1, sin_values, label='Sine Curve', color='blue')
ax.plot(x1, tan_values, label='Tangent Curve', color='green')

# Set labels, legend, and y-axis range
ax.set_xlabel('Angle (2*pi)')
ax.set_ylabel('Values')
ax.set_title('Trigonometric Functions')
ax.legend()
ax.set_ylim(0, 1)  # Set y-axis range from 0 to 1

# Show the plot
plt.show()
```

```python
# For sine (sin):

# sin(0) = 0: The sine of 0 degrees (or 0 radians) is 0.
# sin(/6) = 0.5: The sine of /6 radians (or 30 degrees) is 0.5.
# sin(/4) = 2/2  0.707: The sine of /4 radians (or 45 degrees) is approximately 2/2, which is approximately 0.707.
# sin(/3) = 3/2  0.866: The sine of /3 radians (or 60 degrees) is approximately 3/2, which is approximately 0.866.
# sin(/2) = 1: The sine of /2 radians (or 90 degrees) is 1.

# For cosine (cos):

# cos(2) = 1: The cosine of 2 radians (or 360 degrees) is 1.
# cos() = -1: The cosine of  radians (or 180 degrees) is -1.
# cos(0) = 1: The cosine of 0 degrees (or 0 radians) is 1.
# cos(/6) = 3/2  0.866: The cosine of /6 radians (or 30 degrees) is approximately 3/2, which is approximately 0.866.
# cos(/4) = 2/2  0.707: The cosine of /4 radians (or 45 degrees) is approximately 2/2, which is approximately 0.707.
# cos(/3) = 0.5: The cosine of /3 radians (or 60 degrees) is 0.5.
# cos(/2) = 0: The cosine of /2 radians (or 90 degrees) is 0.

#For tangent (tan):

# tan(0) = 0: The tangent of 0 degrees (or 0 radians) is 0.
# tan(/6) = 3/3  0.577: The tangent of /6 radians (or 30 degrees) is approximately 3/3, which is approximately 0.577.
# tan(/4) = 1: The tangent of /4 radians (or 45 degrees) is 1.
# tan(/3) = 3: The tangent of /3 radians (or 60 degrees) is 3.
# tan(/2) is undefined: The tangent of /2 radians (or 90 degrees) is undefined, as it results in division by zero.
```

# Histogram with fitted normal distribution

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some example data (exam scores)
np.random.seed(42)
exam_scores = np.random.randint(50, 100, 100)  # Generate 100 random scores between 50 and 100

# Create a histogram using plt.hist()
plt.hist(exam_scores, bins=10, edgecolor='black', color='blue', density=True)  

# 'bins' specifies the number of bins or intervals

# density=True: When density is set to True, the heights of the histogram bars are scaled 
# so that the area under the entire histogram is equal to 1. This creates a 
# normalized probability density distribution, which allows you to directly interpret the histogram 
# in terms of probabilities. The heights of the bars represent relative frequencies or probabilities 
# of the data falling within each bin.

# Fit a normal distribution using numpy's mean and standard deviation functions
mu = np.mean(exam_scores)
std = np.std(exam_scores)
x = np.linspace(min(exam_scores), max(exam_scores), 100)
pdf = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / std)**2)
plt.plot(x, pdf, 'r', label=f'Fit Line: $\mu={mu:.2f}$, $\sigma={std:.2f}$')

##############################
# Probability Distribution Function : PDF(x) = (1 / ( * (2 * ))) * e^(-0.5 * ((x - ) / )^2)

# 1 / ( * (2 * )) This equation is responsible for scaling the height of the normal distribution curve 
# in a way that ensures the total area under the curve equals 1, making it a valid probability density function.

# : Mean of the distribution.
# : Standard deviation of the distribution.
# : Pi (approximately 3.14159).
# e: Euler's number (approximately 2.71828).

# 1 / (std * np.sqrt(2 * np.pi))

# std: Represents the standard deviation of the normal distribution. 
# It quantifies how spread out the data points are around the mean.
# np.pi: The mathematical constant pi, approximately 3.14159.
# np.sqrt(): The square root function.

# Gaussian(x) = e^(-0.5 * ((x - ) / )^2)

# This equation shapes the curve of the normal distribution, centered at the mean and symmetrically
# decreasing away from it.

# np.exp(-0.5 * ((x - mu) / std)**2

# mu: This is the mean (average) of the normal distribution. It represents the center of the curve. 
# When (x - mu) is positive, it means x is to the right of the mean; when it's negative, 
# it means x is to the left of the mean.

# The exponentiated term (-0.5 * ((x - mu) / std)**2) is negative, and raising e to a negative power results
# in values that decrease as you move away from the mean.
###########################

###########################

# Let's break down the line plt.plot(x, pdf, 'r', label=f'Fit Line: $\mu={mu:.2f}$, $\sigma={std:.2f}$') step by step:

# plt.plot(x, pdf, 'r', ...): This part of the code uses the plt.plot() function to create a line plot.

# x: The x-values where you want to plot the curve. These are usually evenly spaced points that cover 
# the range of the data or the curve you want to visualize.
# pdf: The y-values of the curve. In this case, it's the probability density values calculated using 
# the normal distribution's PDF formula.
# 'r': This specifies the color of the line. 'r' stands for red. You can use other color codes as well, 
# such as 'b' for blue, 'g' for green, etc.
# label=f'Fit Line: $\mu={mu:.2f}$, $\sigma={std:.2f}$': This part of the code sets a label for the plot line. 
# The label will be used in the legend to identify the line.

# label=: This specifies the label text for the plot line.
# f'...': This is an f-string, which allows you to embed variable values and formatted text within a string.
# $\mu={mu:.2f}$: This part of the f-string embeds the value of mu (mean) with 
# 2 decimal places using .2f format specifier. The $...$ indicates LaTeX-style formatting,
# which is often used in math and science to format equations and symbols.
# $\sigma={std:.2f}$: This part of the f-string embeds the value of std (standard deviation) with 
# 2 decimal places using .2f format specifier.
# In summary, the plt.plot() line creates a line plot using the x and pdf values, setting the line color to red.
# It also adds a label to the plot line that includes the mean (mu) and standard deviation (std) values formatted 
# with 2 decimal places using LaTeX-style formatting. This label will later be used in the legend to provide 
# information about the fitted line on the plot.

#############################


# Add labels and title
plt.xlabel('Exam Scores')
plt.ylabel('Probability Density')
plt.title('Histogram of Exam Scores with Fitted Normal Distribution')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

# Line of best fit 

### Using Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
students_gym = [130, 140, 150, 130, 90, 150, 140, 180, 170, 160, 100, 100]

# Create a scatter plot
plt.scatter(months, students_gym, color='red')

# Fit a linear regression line
coefficients = np.polyfit(range(len(months)), students_gym, 1)
line = np.polyval(coefficients, range(len(months)))

# Add the best-fit line
plt.plot(months, line, label='Best Fit Line', color='blue')

# Add labels and title
plt.xlabel('Months')
plt.ylabel('Number of Students')
plt.title('Number of Students Going to College Gymnasium Each Month')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

## In this code, the np.polyfit() function fits a linear regression line to the data, and np.polyval() evaluates the line's values. The plt.plot() function is used to plot the best-fit line along with the scatter plot. A legend is also added to the plot to indicate the best-fit line.

### Using least square method 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Data
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
students_gym = [130, 140, 150, 130, 90, 150, 140, 180, 170, 160, 100, 100]

# Create a scatter plot
plt.scatter(months, students_gym, color='red')

# Fit a linear regression line using least squares
slope, intercept, r_value, p_value, std_err = linregress(range(len(months)), students_gym)
line = slope * np.array(range(len(months))) + intercept

# Add the best-fit line
plt.plot(months, line, label='Best Fit Line', color='blue')

# Add labels and title
plt.xlabel('Months')
plt.ylabel('Number of Students')
plt.title('Number of Students Going to College Gymnasium Each Month')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

### To fit a linear regression line using the least squares method, we can use the scipy.stats.linregress() function to calculate the slope, intercept, and other statistics of the linear regression line. 

# Box Plot 

```python
# https://www.simplypsychology.org/boxplots.html
```

### In a box plot, also known as a box-and-whisker plot, the various components represent different statistical measures of a dataset's distribution. Here's what each part of the box plot represents:

### Top of the Plot: The top line of the box represents the upper quartile (Q3), which is the value below which 75% of the data falls. In other words, 25% of the data points are above this line.

### Bottom of the Plot: The bottom line of the box represents the lower quartile (Q1), which is the value below which 25% of the data falls. Thus, 75% of the data points are above this line.

### Box: The box itself represents the interquartile range (IQR), which is the range between the upper and lower quartiles (Q3 - Q1). This range contains the middle 50% of the data.

### Yellow Line (Median): Inside the box, there's a horizontal line which represents the median. The median is the value that separates the data into two halves: 50% of the data points are above the median, and 50% are below it.

### Whiskers: The whiskers extend from the top and bottom of the box to represent the spread of the data. The exact definition of the whisker length can vary depending on the method used to create the plot. Generally, they extend to a certain distance (usually 1.5 times the IQR) from the quartiles. Data points beyond the whiskers are considered potential outliers and are typically plotted as individual points.

### Outliers: Data points that fall outside the whiskers are considered outliers and are often plotted as individual points.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

#datetime and timedelta from the datetime module: These are used to work with dates and time intervals.


# Generate example time series data
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 12, 31)
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# Here, you're defining a time period from January 1, 2023, to December 31, 2023. 
# Then, you're using pd.date_range() to generate a sequence of dates (date_range) within 
# this time period with a frequency of one day ('D').

data_values = np.random.randn(len(date_range)).cumsum()

# This line generates a NumPy array of random numbers using the normal distribution (np.random.randn()) 
# with a length equal to the length of the date_range. The .cumsum() function is used to compute the 
# cumulative sum of these random numbers, effectively creating a random walk-like time series data.

# Create a DataFrame
df = pd.DataFrame({'Date': date_range, 'Value': data_values})

# You're creating a Pandas DataFrame named df that consists of two columns: 'Date' and 'Value'. 
# The 'Date' column contains the date_range, and the 'Value' column contains the cumulative sum of random numbers,
# representing the time series data.

# Set the Date column as the index
df.set_index('Date', inplace=True)

# This line sets the 'Date' column as the index of the DataFrame. 
# This is typical for time series data because it allows for convenient time-based indexing and plotting.

# Create a time series plot
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Value'], label='Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Plot')
plt.legend()
plt.grid()
plt.show()

# plt.figure(figsize=(10, 6)): This sets the size of the figure.
# plt.plot(df.index, df['Value'], label='Time Series Data'): 
# This line plots the time series data, where df.index represents the dates on the x-axis and df['Value'] 
# represents the corresponding values on the y-axis. The label parameter provides a label for the legend.
# plt.xlabel('Date') and plt.ylabel('Value'): These lines set the labels for the x-axis and y-axis, respectively.
# plt.title('Time Series Plot'): This sets the title of the plot.
# plt.legend(): This displays the legend.
# plt.grid(): This adds grid lines to the plot.
# plt.show(): This displays the entire plot.
```

# The purpose of the code is to generate and visualize a 3D plot of a parabolic surface in a 3D coordinate space.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# In this part of the code, we are importing the required libraries:

# numpy as np: A library for numerical computations.
# matplotlib.pyplot as plt: A plotting library used to create various types of visualizations.
# mpl_toolkits.mplot3d as Axes3D: A toolkit within Matplotlib that enables the creation of 3D plots.

# Create data for the 3D plot
u = np.linspace(-5, 5, 100)
v = np.linspace(-5, 5, 100)
u, v = np.meshgrid(u, v)

# Here, we are generating a grid of values using np.linspace() to create a range of values for u and v. 
# Then, we are using np.meshgrid() to create coordinate matrices u and v from these ranges. 
# These coordinate matrices will be used to define the coordinates in the x-y plane for the 3D plot.

x = u
y = v
z = u**2 + v**2

# Here, you're defining the coordinates for the 3D surface plot. x, y, and z 
# represent the x, y, and z coordinates of the points on the surface. 
# The equation z = u**2 + v**2 represents a parabolic surface in 3D space.

# Create a figure and a 3D subplot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Here, you're creating a figure and a 3D subplot using add_subplot() with projection='3d'. 
# This prepares the figure for the 3D visualization.

# Plot the 3D surface
surf = ax.plot_surface(x, y, z, cmap='viridis')

# This line of code plots the 3D surface using the plot_surface() function. 
# It takes the x, y, and z coordinates, and the cmap='viridis' argument specifies 
# the colormap used to color the surface.

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')  
ax.set_title('3D Surface Plot')

# Here, we are adding labels and a title to the 3D plot. set_xlabel(), set_ylabel(), and set_zlabel() 
# set labels for the x, y, and z axes, respectively. set_title() sets the title of the plot.

# Display all the axes
ax.set_box_aspect([1, 1, 1])  # Adjust aspect ratio for Z-axis visibility

# This line adjusts the aspect ratio of the plot to ensure that all three axes have the same scale. 
# This helps with the visibility of the plot in 3D space.

# Set plot limits to ensure visibility
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(0, 50)

# These lines set the limits for the x, y, and z axes to control the visible range of the plot.

# Add gridlines on all three axes
ax.grid(True)

# This line adds gridlines to the plot on all three axes.

# Add padding around the plot
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)

# This adjusts the spacing around the plot using subplots_adjust(), ensuring that the plot fits well within the figure.

# Show the plot
plt.show()

# Finally, this line displays the plot.
```
